```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('ggplot')
```

```python
def f1(n):
    return np.log(n)

def f2(n):
    return n

def f3(n):
    return n**2

def f4(n):
    return n**3
    
def f5(n,k):
    return n**k

n = np.arange(1, 8)

plt.figure(figsize=(10, 8))
plt.ylim(0, 120)
plt.plot(n, f1(n), c='blue')
plt.plot(n, f2(n), c='red')
plt.plot(n, f3(n), c='green')
plt.plot(n, f4(n), c='purple')
plt.plot(n, f5(n,4), c='yellow')
plt.xlabel('Size of input (n)', fontsize=16)
plt.ylabel('Number of operations', fontsize=16)
plt.legend(['$\mathcal{O}( \log n)$', '$\mathcal{O}( n)$', '$\mathcal{O}(n^2)$',  '$\mathcal{O}(n^3)$', '$\mathcal{O}(n^k)$'], loc='best', fontsize=12)
plt.show()
```

Big-O notation doesn't directly determine whether an algorithm is effective.
From the above notations if we want to consider the most effective ones, then logarithmic or linear will be better than others. For example, a O(N log N) or O(N) algorithm would be a really effective algorithm for some large tasks, such as sorting or searching. And, even best, is binary search, who runs in O(log N), because it divides the search in half every time a number is checked (because all the bigger/smaller values are discarded). Basically time complexity totally depends on the problems we need to solve. Comparing the execution time of algorithms, we can give a conclusion from the graph is that,
**polynomial**: very bad; 
**cubic**: bad, avoid if possible (For large datasets, it takes a very long time to execute);
**quadratic**: bad, avoid if possible (In most scenarios and particularly for large data sets, algorithms with quadratic time complexities take a lot of time to execute);
**linear**: good (here time complexity grows in direct proportion to the size of the input)
**logarithmic**: great (Algorithms with this complexity make computation amazingly fast.)

O(logn)<O(n)<O(n^2)<O(n^3)<O(n^k)
