```python
train=pd.read_csv('/content/drive/My Drive/train.csv')
```

```python
train = train[['target','comment_text']]
train.head()
```

```python
x = train['comment_text'].tolist()
x[:5]
```

```python
y = train['target'].tolist()
y
```

```python
from keras.preprocessing.text import text_to_word_sequence
from keras.preprocessing.text import Tokenizer  
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import LSTM, Bidirectional, GRU, Dropout,Embedding ,Dense
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
vocab_sz = 10000
maxlen=100
x_train, x_test, y_train, y_test = train_test_split(x[:60000], y[:60000], test_size=0.3, random_state=42)
tok = Tokenizer(num_words=vocab_sz, oov_token='UNK')
tok.fit_on_texts(x)
x_train = tok.texts_to_sequences(x_train)
x_test = tok.texts_to_sequences(x_test )
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
```

```python
model = Sequential()
model.add(Embedding(vocab_sz+1, 50, input_length=maxlen))
model.add(LSTM(128))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
y_train= np.array (y_train)
y_test= np.array(y_test)
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x_train, y_train,batch_size=64, epochs=10
          ,validation_data=(x_test, y_test))
```

```python
test= test['comment_text']
test.head()
```

```python
xt = test.tolist()
xt[:5]
```

```python
vocab_sz = 10000
maxlen=100
tok = Tokenizer(num_words=vocab_sz, oov_token='UNK') #oov_token replace any unknown words with a token of our choosing
tok.fit_on_texts(xt)
xt = tok.texts_to_sequences(xt)
xt = pad_sequences(xt, maxlen=maxlen) #to ensure all the sequences to have same length
print(xt)
```

```python
y_pred= model.predict(xt)
```

```python
y_pred = np.where(y_pred>=0.5,1.,0.)
submission = pd.read_csv('/content/drive/My Drive/sample_submission_second.csv')
submission['prediction'] = y_pred
#submission_df.to_csv('submission.csv',index = False)
submission.head()
```
